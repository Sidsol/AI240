## Exercise 4 -  Draw Histogram


```python
# Generate random data
data = np.random.randn(500)
# Create a histogram with density and custom color
plt.hist(data, bins=20, density=True, color='purple',
edgecolor='black', alpha=0.7)
plt.xlabel('Values')
plt.ylabel('Probability Density')
plt.title('Customized Histogram with Density')
plt.show()
```

## Exercise 5 -  Draw Piechart



```python
y = np.array([10, 3, 15, 5])
mylabels = ["Wolves", "Lions", "Penguins", "Elephants"]
plt.pie(y, labels = mylabels)
plt.show()
```
